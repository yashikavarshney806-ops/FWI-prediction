```python
# prepare the dataframe
dataset=pd.DataFrame(california.data,columns=california.feature_names)
dataset.head()
```

```python
# independent and dependent features
df=dataset 
df.head(3)
```

```python
X=df.iloc[:,:-1] # independent
y=df.iloc[:,-1]  # dependent
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=10)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(X_train,y_train) # fit the line for training data
# many features are there so plot is not pasible to see the line
print(regression.coef_)

```

```python
print(regression.intercept_)
```

```python
#  prediction for test data
# y_test -- The true labels for the test set. These are not used for prediction — they’re only used to evaluate how good our predictions are.

y_pred=regression.predict(X_test)
y_pred
```

```python
# Performance metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
rmse=np.sqrt(mse)
score=r2_score(y_test,y_pred)
print(mse)
print(mae)
print(rmse)
print(score)
```

```python
# adjusted r2_score
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel("Test Truth Data")
plt.ylabel("Test Predicted Data")
```
